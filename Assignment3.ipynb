```python
import pandas as pd 
from sklearn import preprocessing
```

```python
data=pd.read_csv(r"C:\Users\Welcome\Music\ass3\Book2.csv")
data
```

```python
# Calculate mean, median, and mode 
mean_values = data.mean()
median_values = data.median()
mode_values = data.mode().iloc[0] 
```

```python
print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
```

```python
grouped_age_mean = data.groupby(['Gender'])['Placement_Score'].mean()
print("\nGrouped Age Mean:\n", grouped_age_mean)
```

```python
# Encoding categorical variable (Gender)
enc = preprocessing.OneHotEncoder()
enc_data = pd.DataFrame(enc.fit_transform(data[['Gender']]).toarray())
```

```python
# Concatenate encoded values to DataFrame
data_encode = data.join(enc_data)
print("\nEncoded DataFrame:\n", data_encode)
```
